```python
import matplotlib.pyplot as plt
import re 

episodes = []
rewards = []
with open('test.txt', 'r') as f:
    for row in f:
        row = row.split(', ')
        episodes.append(row[1])
        rewards.append(row[0])

num_episodes = []
av_reward = []
for i in rewards:
    i = re.sub(r'[^0-9.]', '', i)
    av_reward.append(i)

for j in episodes:
    j = re.sub(r'[^0-9.]', '', j)
    num_episodes.append(j)
   
plt.plot(num_episodes, av_reward)
plt.xticks(color='w')
plt.yticks(color='w')
plt.xlabel('Number of episodes', fontsize = 12)
plt.ylabel('Average reward', fontsize = 12)
plt.title('Agent Performance', fontsize = 15)
plt.show()
```
