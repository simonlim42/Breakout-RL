```python
import matplotlib.pyplot as plt
import re 

episodes = []
rewards = []
averageQ=[]
maxval=[]
epsilon=[]
with open('dqn.txt', 'r') as f:
    for text in f:
        row = text.split(', ')
        averageQ.append(row[5])
        maxval.append(row[4])
        epsilon.append(row[3])
        episodes.append(row[1])
        rewards.append(row[0])

num_episodes = []
av_reward = []
average_Q=[]
epsilons=[]
maxvals=[]
for i in rewards:
    i = re.sub(r'[^0-9.]', '', i)
    av_reward.append(float(i))

for j in episodes:
    j = re.sub(r'[^0-9.]', '', j)
    num_episodes.append(float(j))
   
for j in epsilon:
    j = re.sub(r'[^0-9.]', '', j)
    epsilons.append(float(j))
    
for j in maxval:
    j = re.sub(r'[^0-9.]', '', j)
    maxvals.append(float(j))
    
for j in averageQ:
    j = re.sub(r'[^0-9.]', '', j)
    average_Q.append(float(j))
    
plt.plot(num_episodes, av_reward)

plt.scatter(num_episodes,av_reward,sizes=[10])

plt.xlabel('Number of steps', fontsize = 12)
plt.ylabel('Average reward', fontsize = 12)
plt.title('DQNAgent Performance', fontsize = 15)
plt.show()

plt.plot(num_episodes, maxvals)
plt.scatter(num_episodes,maxvals,sizes=[10])
plt.xlabel('Number of steps', fontsize = 12)
plt.ylabel('Max reward', fontsize = 12)
plt.title('DQNAgent Performance', fontsize = 15)
plt.show()


plt.plot(num_episodes, average_Q)
plt.scatter(num_episodes,average_Q,sizes=[10])
plt.xlabel('Number of steps', fontsize = 12)
plt.ylabel('Average Q', fontsize = 12)
plt.title('DQNAgent Performance', fontsize = 15)
plt.show()
```
